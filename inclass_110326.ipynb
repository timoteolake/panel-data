```python
import numpy as np
import pandas as pd

from linearmodels import PooledOLS          # Pooled model
from linearmodels import RandomEffects      # Random-effect model
from linearmodels import PanelOLS           # Fixed-effect model
from linearmodels import FirstDifferenceOLS # First difference model

from linearmodels.panel import compare      # Compare the results of multiple models
from statsmodels.api import add_constant    # for matrices of regression design
```

```python
df = pd.read_csv('Gasoline.csv')
df.head()
```

```python
panel_df = df.set_index(['country', 'year'])
panel_df.head()
```

```python
mod_pl = PooledOLS.from_formula(formula='lgaspcar~1+lincomep+lrpmg+lcarpcap', data=panel_df)
res_pl = mod_pl.fit()
res_pl
```

```python
mod_re = RandomEffects.from_formula(formula='lgaspcar~1+lincomep+lrpmg+lcarpcap', data=panel_df)
res_re = mod_re.fit()
res_re
```

```python
mod_fe = PanelOLS.from_formula(formula='lgaspcar~1+lincomep+lrpmg+lcarpcap+EntityEffects', data=panel_df, drop_absorbed=True)
res_fe = mod_fe.fit()
res_fe
```

```python
mod_fd = FirstDifferenceOLS.from_formula(formula='lgaspcar~lincomep+lrpmg+lcarpcap', data=panel_df)
res_fd = mod_fd.fit()
res_fd
```

```python
df = pd.read_csv('Wages.csv')
df.head()
```

```python
panel_df = df.set_index(['id', 'time'])
panel_df.head()
```

```python
mod_pl = PooledOLS.from_formula(formula='lwage~1+ed+exp+I(exp**2)+wks', data=panel_df)
res_pl = mod_pl.fit()
res_pl.params.round(3)
```

```python
mod_re = RandomEffects.from_formula(formula='lwage~1+ed+exp+I(exp**2)+wks', data=panel_df)
res_re = mod_re.fit()
res_re.params.round(3)
```

```python
mod_fe = PanelOLS.from_formula(formula='lwage~1+ed+exp+I(exp**2)+wks+EntityEffects', data=panel_df, drop_absorbed=True)
res_fe = mod_fe.fit()
res_fe.params.round(3)
```

```python
mod_pl = PooledOLS.from_formula(formula='lwage~1+ed+exp+I(exp**2)+wks+south+smsa', data=panel_df)
res_pl = mod_pl.fit()
res_pl.params.round(3)
```

```python
mod_re = RandomEffects.from_formula(formula='lwage~1+ed+exp+I(exp**2)+wks+south+smsa', data=panel_df)
res_re = mod_re.fit()
res_re.params.round(3)
```

```python
mod_fe = PanelOLS.from_formula(formula='lwage~1+ed+exp+I(exp**2)+wks+south+smsa+EntityEffects', data=panel_df, drop_absorbed=True)
res_fe = mod_fe.fit()
res_fe.params.round(3)
```
